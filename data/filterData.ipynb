```python
import pandas as pd
import numpy as np
```

```python
# Load the CSV files
df1 = pd.read_csv('OECD_GDP.csv')
df2 = pd.read_csv('worldBank_GDP.csv')
# Select and rename columns from the first DataFrame
df1_selected = df1[['COU', 'Country', 'YEA', 'GDP']]
```

```python
# Combine the DataFrames
df2.replace('..', np.nan, inplace=True)
df2.dropna(inplace=True)
combined_df = pd.concat([df1_selected, df2])
```

```python
# Save the combined DataFrame to a new CSV file
combined_df.to_csv('GDP.csv', index=False)
```

```python
df4 = pd.read_csv('employee_activity.csv')
df4.head()
```

```python
# df1_selected = df1[['COU', 'Country', 'YEA', 'GDP']]
employee_sector = df4[['LOCATION', 'FREQUENCY', 'Subject','TIME', 'Value']]
employee_sector.head()
```

```python
import re
def get_sector_code(x):
    res = re.findall(r'\(.*?\)', x)
    code = res[-1].strip('(').strip(')')
    code = code.replace('ISIC rev.4,', '')
    code = code.replace('ISIC rev4,', '')
    return code

employee_sector['sector_code'] = employee_sector['Subject'].apply(get_sector_code)
employee_sector.head()
```

```python
test_string = "Employees in Agriculture,hunting and forestry (ISIC rev4, A)";
re.sub("[\(\[].*?[\)\]]", "", test_string).rstrip(' ')
```

```python
def remove_code(x):
    subject = re.sub("[\(\[].*?[\)\]]", "", x)
    return subject.rstrip(' ')

employee_sector['Subject'] = employee_sector['Subject'].apply(remove_code)
employee_sector.head()
```

```python
employee_sector.to_csv('employee_sector.csv', index=False)
```
